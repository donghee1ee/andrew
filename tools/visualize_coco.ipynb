```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def draw_yolo_bbox(image, bbox):
    """
    Draw YOLO bounding box on the image.
    
    :param image: The image (as a numpy array) on which to draw.
    :param bbox: The bounding box in YOLO format [center_x, center_y, width, height] (normalized).
    """
    img_height, img_width = image.shape[:2]

    # Denormalize the coordinates
    center_x, center_y, width, height = bbox
    center_x *= img_width
    center_y *= img_height
    width *= img_width
    height *= img_height

    # Convert to top-left corner format
    x_min = int(center_x - width / 2)
    y_min = int(center_y - height / 2)

    # Draw the bounding box
    cv2.rectangle(image, (x_min, y_min), (x_min + int(width), y_min + int(height)), (255, 0, 0), 2)

    # Display the image
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
def draw_coco_bbox(image, bbox):
    """
    Draw COCO bounding box on the image.
    
    :param image: The image (as a numpy array) on which to draw.
    :param bbox: The bounding box in COCO format [x_min, y_min, width, height].
    """
    x_min, y_min, bbox_width, bbox_height = bbox

    # Draw the bounding box
    cv2.rectangle(image, (int(x_min), int(y_min)), 
                        (int(x_min + bbox_width), int(y_min + bbox_height)), 
                        (0, 255, 0), 2)

    # Display the image
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
def draw_yolo_bbox(image, bbox):
    """
    Draw YOLO bounding box on the image.
    
    :param image: The image (as a numpy array) on which to draw.
    :param bbox: The bounding box in YOLO format [center_x, center_y, width, height] (normalized).
    """
    image = np.array(image)
    img_height, img_width = image.shape[:2]

    # Denormalize the coordinates
    center_x, center_y, width, height = bbox
    center_x *= img_width
    center_y *= img_height
    width *= img_width
    height *= img_height

    # Convert to top-left corner format
    x_min = int(center_x - width / 2)
    y_min = int(center_y - height / 2)


    # Draw the bounding box
    cv2.rectangle(image, (x_min, y_min), (x_min + int(width), y_min + int(height)), (255, 0, 0), 2)

    # Display the image
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
def draw_bbox(image, bbox):
    """
    Draw a bounding box on the image.

    :param image: The image (as a numpy array) on which to draw.
    :param bbox: The bounding box in the format [x_min, y_min, x_max, y_max] in pixels.
    """
    x_min, y_min, x_max, y_max = bbox

    # Draw the bounding box
    cv2.rectangle(image, (int(x_min), int(y_min)), (int(x_max), int(y_max)), (0, 0, 255), 2)

    # Display the image
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
image = cv2.imread('image-3.jpg')

# Example YOLO bbox (normalized): [center_x, center_y, width, height]
bbox = [0.7564609375, 0.5630520833333333, 0.8922343749999999, 0.9041041666666667] # Example values

draw_yolo_bbox(image, bbox)
```
